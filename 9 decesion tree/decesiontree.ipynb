# Target Feature is the DRUG 

# Step 3 =  seperate x and y

```python
df["Drug"].unique()
```

### more than two classes so multi class classification problem

```python
X = df.drop(columns=["Drug"])
Y = df[["Drug"]]
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts().plot(kind="bar")
```

```python
Y.value_counts()
```

```python
cat = list(X.columns[X.dtypes == "object"])
con = list(X.columns[X.dtypes != "object"])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy="median"),
    StandardScaler()
)
```

```python
cat_pipe = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore",sparse_output=False,drop = "first")
)
```

```python
pre = ColumnTransformer(
    [
        ("num",num_pipe,con),
        ("cat",cat_pipe,cat),
    ]
).set_output(transform="pandas")
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

# step 5 

```python
from sklearn.model_selection import train_test_split

xtrain , xtest ,ytrain ,ytest = train_test_split(X_pre,Y,test_size=0.4,random_state=42)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
xtest.head()
```

# step 6 = model building

```python
from sklearn.tree import DecisionTreeClassifier   # classifier coz cat is result
```

```python
model = DecisionTreeClassifier(
    max_depth=1,
    min_samples_split=5,
    min_samples_leaf=5,
    criterion="entropy"
)
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
model.score(xtrain,ytrain)
```

# Hyperparameters tuning

```python
params = {
    "max_depth" : [2,3,4,5,6] ,
    "min_samples_split" : [5,6,7,8] ,
    "min_samples_leaf" : [5,6,7,8],
    "criterion" : ["gini", "entropy"]
}
```

```python
from sklearn.model_selection import GridSearchCV

base_dtc = DecisionTreeClassifier()

gscv = GridSearchCV(
    estimator=base_dtc,
    param_grid= params,
    scoring="f1_macro",
    cv=5,
    n_jobs=-1
)

gscv.fit(xtrain, ytrain)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python

best_dtc = gscv.best_estimator_
best_dtc
```

# Step 7 - Evalute the model
