# 1 = DATA INGESTION


```python
import pandas as pd
df = pd.read_csv("training_set.csv",na_values=["","NA"],keep_default_na=False)
df.head()
```

# 3 = sep x and y

```python
X = df.drop(columns=["Id","SalePrice"])
Y = df[["SalePrice"]]
```

```python
X.head()
```

```python
Y.head()
```

# 4 = Feature selection

```python
cat = list(X.columns[X.dtypes=="object"])
con = list(X.columns[X.dtypes!="object"])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy="mean"),
    StandardScaler()
)
```

```python
cat_pipe = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OrdinalEncoder()
)
```

```python
pre1 = ColumnTransformer(
    [
        ("num",num_pipe,con),
        ("cat",cat_pipe,cat)
    ]
).set_output(transform="pandas")
```

```python
pre1
```

```python
X_pre = pre1.fit_transform(X)
X_pre
```

# 5 = Apply feature selection

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
base_model = LinearRegression()
```

```python
sel = SequentialFeatureSelector(base_model, n_features_to_select="auto", direction="forward", n_jobs=-1)
sel.fit(X_pre, Y)
imp_cols = sel.get_feature_names_out()
imp_cols
```

```python
imp_cols.shape
```

```python
len(imp_cols)
```

```python

sel_cols = [col.split("__")[-1] for col in imp_cols]
sel_cols
```

```python
X_sel = X[sel_cols]
X_sel
```

# Feature selection is done

# Create a final pipeline of selected features


```python
cat_sel = list(X_sel.columns[X_sel.dtypes == "object"])
con_sel = list(X_sel.columns[X_sel.dtypes != "object"])
```

```python
cat_sel
```

```python
con_sel
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pipe2 = make_pipeline(
    SimpleImputer(strategy="mean"),
    StandardScaler()
)
```

```python
cat_pipe2 = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore",sparse_output=False,drop="first")
)
```

```python
pre2 = ColumnTransformer(
    [
        ("num", num_pipe2, con_sel),
        ("cat", cat_pipe2, cat_sel)
    ]
).set_output(transform="pandas")
```

```python
pre2
```

```python
X_sel_pre = pre2.fit_transform(X_sel)
X_sel_pre.head()
```

# Apply Train Test Split


```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X_sel_pre,Y,test_size=0.2,random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
model = LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
from sklearn.metrics import(
    root_mean_squared_error,mean_absolute_error,r2_score,mean_absolute_percentage_error
)
```

```python
def evaluate_model(model, x, y):
    ypred = model.predict(x)
    rmse = root_mean_squared_error(y, ypred)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)
    print(f"RMSE : {rmse:.2f}")
    print(f"MAE : {mae:.2f}")
    print(f"MAPE : {mape:.2%}")
    print(f"R2 : {r2:.2%}")
```

```python
evaluate_model(model,xtrain,ytrain)
```

```python
evaluate_model(model,xtest,ytest)
```

# score less than 80 so not a good model
